# 三、user_info.csv

```python
user_info = user_log[['user_id','age_range','gender']]
user_info = pd.DataFrame(user_info)
user_info = user_info.drop_duplicates() 
user_info.reset_index(drop=True,inplace=True)   # 重置索引
user_info.to_csv('user_1000.csv',index=None)
user_info # 1000 
```

```python
item_info = user_log[['item_id','seller_id','brand_id','cat_id']]
item_info = pd.DataFrame(item_info)
item_info = item_info.drop_duplicates() 
item_count = item_info['item_id'].value_counts().reset_index()
item_count = item_count.rename(columns={'index':'item_id','item_id':'count'})
item_list = list(item_count['item_id'][item_count['count']==1].values) # 59671 
item_info =  item_info[item_info['item_id'].isin(item_list)]
item_info.reset_index(drop=True,inplace=True)   # 重置索引
item_info.to_csv('item_info_1000.csv',index=None)
item_info # 278426   
```

```python
user_log = user_log[user_log['item_id'].isin(item_list)]
user_log.drop_duplicates()  # 去重 
user_log.reset_index(drop=True,inplace=True)   # 重置索引
user_log.to_csv('user_log_1000.csv',index=None) # 存入csv文件并删除索引
user_log # 154006     按用户名、时间排序
```
